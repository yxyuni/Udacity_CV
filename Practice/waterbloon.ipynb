```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('water_bloons.jpg')
img_copy = np.copy(img)
img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_copy)
```

```python
#RGB
R = img_copy[:, :, 0]
G = img_copy[:, :, 1]
B = img_copy[:, :, 2]
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=[20, 10])
ax1.set_title('Red')
ax1.imshow(R, cmap='gray')
ax2.set_title('Green')
ax2.imshow(G, cmap='gray')
ax3.set_title('Blue')
ax3.imshow(B, cmap='gray')
```

```python
img_copy2 = cv2.cvtColor(img_copy, cv2.COLOR_RGB2HSV)
```

```python
H = img_copy2[:, :, 0]
S = img_copy2[:, :, 1]
V = img_copy2[:, :, 2]
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=[20, 10])
ax1.set_title('Hue')
ax1.imshow(H, cmap='gray')
ax2.set_title('Satuation')
ax2.imshow(S, cmap='gray')
ax3.set_title('Value')
ax3.imshow(V, cmap='gray')
```

```python
low_pink = np.array([180, 0, 100])
high_pink = np.array([255, 255, 230])
mask = cv2.inRange(img_copy, low_pink, high_pink)
masked_img = np.copy(img_copy)
masked_img[mask == 0] = [0, 0, 0]
plt.imshow(mask)
plt.figure()
plt.imshow(masked_img)
```

```python
low_h = np.array([160, 0, 0])
high_h = np.array([180, 255, 255])
mask_h = cv2.inRange(img_copy2, low_h, high_h)
mask_hsv = np.copy(img_copy)
mask_hsv[mask_h == 0] = 0
plt.imshow(mask_hsv)
```
